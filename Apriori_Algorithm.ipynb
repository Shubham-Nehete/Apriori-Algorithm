### Importing Libraries

```python
import pandas as pd
import numpy as np
from apyori import apriori
```

### Importing Dataset

```python
dataset = pd.read_csv('Market_Basket_Optimisation.csv', header= None)
dataset.head()
```

```python
dataset.shape
```

```python
transactions = []
for i in range(0,7501):
    transactions.append([str(dataset.values[i,j]) for j in  range(0,20)])
```

```python
transactions
```

### Trainig the Apriori model on the dataset

```python
rules = apriori(transactions, min_support = 0.003, min_confidence = 0.5, min_lift = 3, min_length = 2)
```

### Visulaising the results

```python
results = list(rules)
```

```python
for i in results:
    print('\n')
    print(i)
    print("--------------------------------------------------------------------------------------------------------------------------")
```

```python
df_results = pd.DataFrame(results)
```

```python
df_results
```

```python
df_results['ordered_statistics'][0][0]
```

```python
# Segregating the "ordered_statistics" column in order to get desirable results.
value_1 = []
value_2 = []
value_3 = []
value_4 = []

# loop by number of rows and append 1 by 1 value in a separate list.
for i in range(df_results.shape[0]):
    res = df_results['ordered_statistics'][i][0]
    value_1.append(list(res[0]))
    value_2.append(list(res[1]))
    value_3.append(res[2])
    value_4.append(res[3])
```

```python
# Keeping support values separate
value_5 = df_results['support']
```

```python
p = pd.DataFrame(value_1)
q = pd.DataFrame(value_2)
confidence = pd.DataFrame(value_3, columns=['Confidence'])
lift = pd.DataFrame(value_4,columns=['Lift'])
support = pd.DataFrame(value_5)
```

```python
df = pd.concat([p, q, support, confidence, lift], axis= 1)
df.head()
```

```python
df.columns = [0,1,2,3,4,'Support','Confidence','Lift']
```

```python
df.fillna(value= "", inplace= True)
df.head()
```

```python
# Separating Base items
df["Items_base"] = df[0]+str(", ") + df[1]+str(", ") + df[2]
```

```python
# Separating Add items
df[3] = df[3].replace("nan", "")
df[4] = df[4].replace("nan", "")
df['Items_add'] = df[3]+str(", ") + df[4]
```

```python
df.drop([0,1,2,3,4], axis=1, inplace=True)
```

```python
# Just doing some cleaning like removing "," if there is no item in the last.
res1 = []
for x in df['Items_base']:
    if x[-2:] == ', ':
        x = x[:-2]
        res1.append(x)    
    else:
        res1.append(x)

res2 = []
for x in df['Items_add']:
    if x[-2:] == ', ':
        x = x[:-2]
        res2.append(x)
    elif x[:2] == ", ":
        x = x[2:]
        res2.append(x)    
    else:
        res2.append(x)

df['Items_base'] = res1
df['Items_add'] = res2
```

# Final Result

```python
df = df[["Items_base", "Items_add", "Support", "Confidence", "Lift"]]
df
```
